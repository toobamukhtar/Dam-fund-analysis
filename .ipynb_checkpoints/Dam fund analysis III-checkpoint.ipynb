# Dam Fund Analysis - Feature Extraction and Regression

In this part we'll move beyond simply analysing the data and will instead move towards trying to make some inferences regarding the data - predictions on when the dam's target value of (the minimum estimate) 1.5 Trillion rupees will be reached.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

data = pd.read_pickle('./individual_contributions.pkl')
```

We'll be running a regression analysis on this data since the target variable, the funds collected, is a continuous variable. Before we are able to run any sort of regression we need to decide what features we should be using for our regression. Moreover, since we are running a regression it is important to also figure out what exactly our target variable should be. Should it be the **cumulative sum** of the amount collected **till** each day, or should it simply be the amount collected **on** each day? 


```python
data.head()
```

```python
new_data = data.groupby(['Date']).sum().cumsum()
new_data.reset_index(level=0, inplace=True)
new_data.head()
```

```python
palette = sns.color_palette("husl", 5)
fig = plt.figure(figsize=(20,10), dpi=80)
# uses index as x axis
ax = sns.lineplot(x = 'Date',y = 'Amount',data=new_data)
plt.title('Cumulative Amount',fontsize = 30)
plt.xlabel('Date',fontsize = 20)
plt.ylabel('Total Count',fontsize = 20)
```

## Feature Extraction 

We have currently have 3 columns, other than the target variable (Amount), Bank, Name and Date. Which do you think should be used as the independent variable in running the regression? 

Ans: Date Variable.

One possible variable we could use is the Date variable, but it can not be used directly since it is a 'Datetime' object. 

We need to think deeply about what sort of variables can be extracted from simply "Date", and how they would be useful in trying to figure out how many funds are being collected on any given day. One good way to go about it would be to try out the regression on many different features and see which one is better.

```python
# Import the appropriate functions from sklearn #
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, StandardScaler, MinMaxScaler
from sklearn.preprocessing import OrdinalEncoder
```

```python
def interval(x):
    if x<=10:
        return 0
    elif x<=20:
        return 1
    else:
        return 2
 
```

```python
# Extract the right features #
new_data['Day_of_week'] = new_data['Date'].dt.dayofweek 
new_data['month'] = new_data['Date'].dt.month 
new_data['day_numeric'] = new_data['Date'].dt.day
new_data['ordinal_date'] = new_data['Date'].values
new_data['interval'] = new_data['day_numeric'].apply(interval)
new_data.head()
```

```python
def convert(column):
    input_value = np.array(column)
    label_encoder = LabelEncoder()
    integer = label_encoder.fit_transform(input_value)
    print('classes',label_encoder.classes_)
    print('input_value',input_value[0:5])
    print('integer_value',integer[0:5])
    return integer
```

```python
new_data['month'] = convert(new_data['month'])
new_data['interval'] = convert(new_data['interval'])
new_data['Day_of_week'] = convert(new_data['Day_of_week'])
```

```python
new_data.head()
```

```python
# We used ordinal encoder to convert ordinal date because the order matters
input_value = np.array(new_data['ordinal_date'])
input_value = input_value.reshape(len(input_value),1)
ordinal_encoder = OrdinalEncoder()
integer = ordinal_encoder.fit_transform(input_value)
print('input_value',input_value[0:5])
print('integer_value',integer[0:5])
new_data['ordinal_date'] = integer
```

```python
def corr_plot(columns):
    correlation_map = np.corrcoef(new_data[columns].values.T)
    sns.set(font_scale=1.5)
    heatmap = sns.heatmap(correlation_map, cbar=True, annot=True, square=True, fmt='.3f', yticklabels=columns.values, xticklabels=columns.values)
    plt.show()
```

We calculated the correlation between the target variable and our 5 features. We can see that the top 3 features which have a high correlation with amount are ordinal_date, month, interval in this order.

```python
correlation = new_data.corr()
columns = correlation.nlargest(4, 'Amount').index
columns
```

```python
corr_plot(columns)
```

We can see from the correlation matrix that only the ordinal date has the hightest correlation with the cumulative amount. The  day of the week and numeric day are very correlated with the cumulative amount.

In the above plot the prediction made by the model is on the x-axis, and the accuracy of the prediction is on the y-axis. The distance from the line at 0 is how bad the prediction was for that value.
Since residula is Observed – Predicted it means that the positive values for the residual (on the y-axis) mean the prediction was too low, and negative values mean the prediction was too high; 0 means the guess was exactly correct.


## Ridge Regression 

```python
palette = sns.color_palette("husl", 5)
reg = RidgeCV(alphas=(400, 450,1000), cv=None)
reg.fit(input_features, target)
pred = reg.predict(input_features)
print('Coefficients:',reg.coef_)
print('R^2: ', reg.score(input_features, target))
print('MSE: ', mean_squared_error(target, pred))

fig, (ax1) = plt.subplots(ncols=1, figsize=(6,6), dpi=80)
fig.suptitle('Ridge Regression')

ax1.scatter(target, pred)
ax1.set(xlabel='Actual Amount', ylabel='Predicted Amount');
ax1.axis('equal');
```

## Lasso Regression

```python
palette = sns.color_palette("husl", 5)
reg = LassoCV(alphas=(10.0,20.0,100.0,1000), cv=2)
reg.fit(input_features, target)
pred = reg.predict(input_features)
print('Coefficients:',reg.coef_)
print('R^2: ', reg.score(input_features, target))
print('MSE: ', mean_squared_error(target, pred))

fig, (ax1) = plt.subplots(ncols=1, figsize=(6,6), dpi=80)
fig.suptitle('Lasso Regression')

ax1.scatter(target, pred)
ax1.set(xlabel='Actual Amount', ylabel='Predicted Amount');
ax1.axis('equal');

```
